### In this notebook analysis of a network from datadet of reactions is performed.

Firstly, the chemistry library used in the first half of the notebook is imported.

```python
from rdkit import Chem
from rdkit.Chem import inchi

```

get_biggest_reactant_and_product takes as input reaction SMILES, splits the reaction and separated the reactants and product which are then sorted and the  biggest ones is kept. The InChI keys are calculated and the first 14 characters are stored. The function returns smiles and inchi keys for the reactant and product.


The text file with reaction SMILES is loaded and one by one the reactions are processed by get_biggest_reactant_and_product and the results are stored in an array. 
    

```python
import csv
import tqdm
reactions = []
problems = 0
def get_biggest_reactant_and_product(rxnsmiles, same):
    r,_,p = rxnsmiles.split(">")
    reactant_smi = sorted(r.split("."),key=lambda x:len(x))[-1]
    product_smi = sorted(p.split("."),key=lambda x:len(x))[-1]
    reactant_inchikey = inchi.MolToInchiKey(Chem.MolFromSmiles(reactant_smi))[:14]
    product_inchikey = inchi.MolToInchiKey(Chem.MolFromSmiles(product_smi))[:14]
    if reactant_inchikey==product_inchikey:
        print("product = reactant. skipping")
    return reactant_inchikey,product_inchikey,reactant_smi,product_smi

with open('reactionSmilesFigShare2025.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in tqdm.tqdm(reader):
        try:
            reactions.append(get_biggest_reactant_and_product(row[0], same))
        except:
            problems = problems + 1
            print("problem for",row[0])
```

```python
print(problems)
```

In the following cell all of the unique reactions are extracted and the edges are computed from them by finding reactions where the product of one reaction is the reactant of a second one.

```python
from collections import defaultdict

#no self-loops
reactions = [r for r in reactions if r[0]!=r[1]]
#deduplicate
reactions_unique = list(dict.fromkeys(reactions))

# Map each reactant_inchi → list of row indices where it appears
reactant_index = defaultdict(list)
for j, (reactant_inchi, _, _, _) in enumerate(reactions_unique):
    reactant_index[reactant_inchi].append(j)

# For each product, look up matching reactants directly
edges = [
    (j, i)
    for i, (_, product_inchi, _, _) in enumerate(tqdm.tqdm(reactions_unique))
    for j in reactant_index.get(product_inchi, [])
]
```

From the edges a directed graph is then constructed, analyzed by its connectivity and a graph of top connected nodes is visualized using netorkx, pyvis and nunmpy.

```python
import networkx as nx

G = nx.DiGraph(edges)
```

```python
import networkx as nx
from collections import Counter
import numpy as np


in_deg  = dict(G.in_degree())
out_deg = dict(G.out_degree())
tot_deg = {n: in_deg[n] + out_deg[n] for n in G.nodes()}

#most connected node
top_node = max(tot_deg, key=tot_deg.get)
print(f"Most connected node : {top_node}")
print(f"  in-degree         : {in_deg[top_node]}")
print(f"  out-degree        : {out_deg[top_node]}")
print(f"  total degree      : {tot_deg[top_node]}")

#top nodes based on their degree
top15 = sorted(tot_deg.items(), key=lambda x: x[1], reverse=True)[:15]
print("\nTop 15 nodes by total degree:")
for node, deg in top15:
    print(f"  {node}  in={in_deg[node]}  out={out_deg[node]}  total={tot_deg[node]}")
for node, deg in top15:
    print(reactions[node])

#top nodes based on their in degree
top15In = sorted(in_deg.items(), key=lambda x: x[1], reverse=True)[:15]
print("\nTop 15 nodes by in degree:")
for node, deg in top15In:
    print(f"  {node}  in={in_deg[node]}  out={out_deg[node]}  total={tot_deg[node]}")    
for node, deg in top15In:
    print(reactions[node])

#top nodes based on their out degree
top15Out = sorted(out_deg.items(), key=lambda x: x[1], reverse=True)[:15]
print("\nTop 15 nodes by out degree:")
for node, deg in top15Out:
    print(f"  {node}  in={in_deg[node]}  out={out_deg[node]}  total={tot_deg[node]}")
for node, deg in top15Out:
    print(reactions[node])

in_vals  = np.array(list(in_deg.values()))
out_vals = np.array(list(out_deg.values()))
tot_vals = np.array(list(tot_deg.values()))

print(f"\nIn-degree  — mean: {in_vals.mean():.2f}  median: {np.median(in_vals):.0f}  max: {in_vals.max()}")
print(f"Out-degree — mean: {out_vals.mean():.2f}  median: {np.median(out_vals):.0f}  max: {out_vals.max()}")
print(f"Total-degree — mean: {tot_vals.mean():.2f}  median: {np.median(tot_vals):.0f}  max: {tot_vals.max()}")

```

```python
THRESHOLD = 85     
h_nodes = [n for n, d in tot_deg.items() if d >= THRESHOLD]
H = G.subgraph(h_nodes)
print(f"\n Subgraph (deg ≥ {THRESHOLD}): {H.number_of_nodes()} nodes, {H.number_of_edges()} edges")

I = nx.DiGraph(list(H.edges))
print(f"\n Subsubgraph: {I.number_of_nodes()} nodes, {I.number_of_edges()} edges")
```

```python
import networkx as nx
from pyvis.network import Network


pos = nx.spring_layout(I, k=3.5, iterations=200, seed=42)
net = Network(height="1200px", width="100%", notebook=True, cdn_resources='remote')
net.from_nx(I)

for node in net.nodes:
    node_id = node["id"]
    x, y = pos[node_id]
    node["x"] = float(x * 2000)
    node["y"] = float(y * 2000)

    deg = I.degree(node_id)
    node["font"] = {"size": 35}
    node["size"] = 5 + deg * 0.3
    node["label"] = str(node_id)
    node["title"] = f"Node: {node_id}<br>Degree: {deg}"
    node["color"] = "#FA9BCF"

for edge in net.edges:
    edge["width"] = 0.5
    edge["color"] = "#658795"

net.toggle_physics(True)

net.show("graphThinEdges.html")
```

In the following cells, from the SMILES of biggest reactant and product the fingerprints are calculated and stored as a numpy array. Same goes for the SMILES array. 

```python
from rdkit.Chem import rdFingerprintGenerator
reactionFPs = []
smis = []

generateFp = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
def fp_arr(smi):
    mol = Chem.MolFromSmiles(smi)
    fp = generateFp.GetFingerprint(mol)
    return np.array(list(fp.ToBitString()), dtype = np.int8)

    
for reaction in reactions:
    fpr = fp_arr(reaction[2])
    fpp = fp_arr(reaction[3])
    cur_reactionFPs = np.concatenate([fpr, fpp])
    reactionFPs.append(cur_reactionFPs)
    smis.append((reaction[2], reaction[3]))

    
X = np.array(reactionFPs)
smiles = np.array(smis)
print(X.shape)
print(smiles.shape)
```
